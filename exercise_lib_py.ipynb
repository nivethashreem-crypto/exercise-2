```python
# Base Class
class Member:
    def __init__(self, name, membershipID):
        self.name = name
        self.membershipID = membershipID

    def display(self):
        print("Name:", self.name)
        print("Membership ID:", self.membershipID)


# Derived Class - Regular Member
class RegularMember(Member):
    def __init__(self, name, membershipID):
        super().__init__(name, membershipID)
        self.borrow_limit = 3

    def getBorrowLimit(self):
        return self.borrow_limit

    # Overriding display()
    def display(self):
        print("\nMember Type: Regular Member")
        print("Name:", self.name)
        print("Borrowing Limit:", self.getBorrowLimit(), "books")


# Derived Class - Premium Member
class PremiumMember(Member):
    def __init__(self, name, membershipID):
        super().__init__(name, membershipID)
        self.borrow_limit = 10
        self.discount = "10% Discount on Membership Fees"

    def getBorrowLimit(self):
        return self.borrow_limit

    # Overriding display()
    def display(self):
        print("\nMember Type: Premium Member")
        print("Name:", self.name)
        print("Borrowing Limit:", self.getBorrowLimit(), "books")
        print("Additional Feature:", self.discount)


# Main Function
if __name__ == "__main__":
    # Create Regular Member Object
    regular = RegularMember("Arun", 101)
    regular.display()

    # Create Premium Member Object
    premium = PremiumMember("Priya", 202)
    premium.display()
```
