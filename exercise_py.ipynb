```python
#base class
class student:
  def __init__(self, name, rollNumber):
    self.name = name
    self.rollNumber = rollNumber
  def display(self):
    print("\nName:", self.name)
    print("Roll Number:", self.rollNumber)


# Derived Class 1
class UndergraduateStudent(student):
    def __init__(self, name, rollNumber, marks):
        super().__init__(name, rollNumber)
        self.marks = marks   # List of 3 subject marks

    def calculateAverage(self):
        average = sum(self.marks) / 3
        print("Average Marks (UG):", average)


# Derived Class 2
class PostgraduateStudent(student):
    def __init__(self, name, rollNumber, marks):
        super().__init__(name, rollNumber)
        self.marks = marks   # List of 4 subject marks

    def calculateAverage(self):
        average = sum(self.marks) / 4
        print("Average Marks (PG):", average)


# Main Program
print("---- Undergraduate Student ----")
name = input("Enter Name: ")
roll = int(input("Enter Roll Number: "))
marks_ug = []

print("Enter marks for 3 subjects:")
for i in range(3):
    mark = float(input(f"Subject {i+1}: "))
    marks_ug.append(mark)

ug = UndergraduateStudent(name, roll, marks_ug)
ug.display()
ug.calculateAverage()


print("\n---- Postgraduate Student ----")
name = input("Enter Name: ")
roll = int(input("Enter Roll Number: "))
marks_pg = []

print("Enter marks for 4 subjects:")
for i in range(4):
    mark = float(input(f"Subject {i+1}:"))
    marks_pg.append(mark)
```
